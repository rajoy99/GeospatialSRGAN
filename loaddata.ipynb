```python
for i in range(0,100):
    fig = plt.figure()
    fig.set_size_inches((5,5))
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.patch.set_facecolor('white')
    fig.add_axes(ax)
    plt.imshow(data1[i])
    plt.savefig('highres_imgs/'+str(i)+'.png')
```
